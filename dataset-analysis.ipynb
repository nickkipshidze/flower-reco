```python
import os
import scipy.io
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
mat = scipy.io.loadmat("./imagelabels.mat")
labels = mat["labels"].squeeze()
```

```python
image_paths = sorted(os.listdir("./flower_images"))
images = []

for path in image_paths:
    images.append(
        Image.open(f"./flower_images/{path}")
    )
```

```python
unique, counts = np.unique(labels, return_counts=True)
print(unique)
print(counts)
```

```python
image_by_class = [[] for _ in range(0, 102)]
for index, image in enumerate(images):
    image_by_class[labels[index]-1].append(image)
```

```python
def preview_images(images, title=None):
    plt.figure(figsize=(5, 5))
    plt.title(title if title else f"Images preview ({len(images)})", fontsize=18)
    plt.axis(False)
    for index, image in enumerate(images):
        plt.subplot(int(np.sqrt(len(images)))+2, int(np.sqrt(len(images))), index+1)
        plt.imshow(image)
        plt.axis(False)
    plt.show()
```
